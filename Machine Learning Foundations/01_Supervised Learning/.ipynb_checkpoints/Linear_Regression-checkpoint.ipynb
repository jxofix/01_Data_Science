# Univariate Linear Regression

Linear regression is a statistical technique used for approximating real values, such as the cost of houses, using one or more continuous variables. It involves establishing the connection between independent and dependent variables by determining the optimal line of best fit. This line, known as the regression line, is represented by a linear equation for one variable x (univariate) in the form: <br>
$f_{(w,b)}(x) = w*x + b $

## Cost function

The most commonly used cost function for linear regression is the Squared error cost function: <br>
$ J(w,b) = \frac{1}{2*m}\ * \sum{(f(x) - y)^{2}} $
, where $ f(x)= \hat{y}\ $ - the estimated value and y is the target value.

Let see and example:

```python
# prepare simple data set
import pandas as pd

df = pd.read_csv('./data/Salary_Data.csv')
print(df.shape)
print(df.head())
```

```python
# lets plot data set using matplotlib
import matplotlib.pyplot as plt

fix, ax = plt.subplots(figsize=(5,4))
ax.set(title='Linear Regression',
       xlabel = 'Years of Experience',
       ylabel = 'Salary')
ax.plot(df.YearsExperience, df.Salary, "xr")
plt.show()
```

### Lets define cost function and plot it for selected variables w and b

```python
import seaborn as sb
import numpy as np
from mpl_toolkits.mplot3d import Axes3D 

def cost_function(x, y, w_i, b_i):
    error = []
    for i in range(0, len(x)):
        error.append((w_i*x[i]+b_i - y[i])**2)

    squared_error = sum(error)/(2*len(x))
    return squared_error

w = np.arange(0, 20000, 100)
b = np.arange(0, 50000, 100)
W, B = np.meshgrid(w, b)
zs = np.array(cost_function(df.YearsExperience, df.Salary, np.ravel(W), np.ravel(B)))
Z = zs.reshape(W.shape)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set (xlabel = 'w',
        ylabel = 'b',
        zlabel = 'J(w,b)'
       )
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.set_zticklabels([])
ax.zaxis.labelpad= -10
ax.plot_surface(W, B, Z, cmap='YlGnBu', edgecolor='none')
ax.set_title('Cost Function')
plt.show()
```

The goal of linear regression is to minimalize cost function min(J(w,b)) -> to find w, b such that the square root error will be minimal. Due to many features models usually have this is performed using Gradient Descent. Gradient descent edits w and b values every iteration (step) closer to the local minimum of J(w, b).

## Gradient descent
for w: <br>
$ w = w - \alpha*\frac{\partial}{\partial w}J(w, b) $ <br><br>
$ w = w - \alpha*\frac{1}{m}\sum_{i=1}^{m}(f(x_{i})-y_{i})*x_{i} \, $
<br><br>
for b:<br>
$ b = b - \alpha*\frac{\partial}{\partial b}J(w, b) $ <br><br>
$ b = b - \alpha*\frac{1}{m}\sum_{i=1}^{m}f(x_{i})-y_{i} \, $


```python
#calculation of gradient (a derivative of cost function J(w,b))
def calc_gradient(x, y, w, b):
    de_w = 0
    de_b = 0

    for i in range(0, len(x)):
        de_w_i = (w * x[i] + b - y[i])*x[i]
        de_b_i = w * x[i] + b - y[i]

        de_w += de_w_i
        de_b += de_b_i

    de_w = de_w / len(x)
    de_b = de_b / len(x)
    
    return de_w, de_b

#updating w, b value to the local minima
def gradient_descent(x, y, w_in, b_in, alpha, num_iter):
    for i in range(num_iter):
        w_in = w_in - alpha * calc_gradient(x, y, w_in, b_in)[0]
        b_in = b_in - alpha * calc_gradient(x, y, w_in, b_in)[1]

        if i % 1000 == 0:
            print(f'Iteration {i:4} | w : {w_in: 8.2f} | b : {b_in: 8.2f}')
    return w_in, b_in
```

```python
# initialize parameters
w_init = 0
b_init = 0

num_iters = 10000
alpha = 1.0e-2

w_fin, b_fin = gradient_descent(df.YearsExperience, df.Salary, w_init, b_init, alpha, num_iters)
print(f'(w,b) found by gradient descent: ({w_fin:8.4f}, {b_fin:8.4f})')
```

### Visualization of Linear regression with obtained w and b

```python
#lets calculate prediction for our Years of experience
y_pred = []
for idx, row in df.iterrows():
    y_pred.append(w_fin*row['YearsExperience']+b_fin)

# visualize
fix, ax = plt.subplots(figsize=(5,4))
ax.set(title='Linear Regression',
       xlabel = 'Years of Experience',
       ylabel = 'Salary')
ax.plot(df.YearsExperience, df.Salary, "xr")
ax.plot(df.YearsExperience, y_pred, 'b')   # added linear regression line
plt.show()
```

### Example how is Linear regression performed using sci-kit learn library

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

X = np.array(df.YearsExperience)
X = X.reshape(-1,1)

lr_model.fit(X, df.Salary)
y_pred = lr_model.predict(X)

fix, ax = plt.subplots(figsize=(5,4))
ax.set(title='Linear Regression',
       xlabel = 'Years of Experience',
       ylabel = 'Salary')
ax.plot(df.YearsExperience, df.Salary, "xr")
ax.plot(df.YearsExperience, y_pred, 'b')
plt.show()
```

### Lets compare w and b values obtained 'manually' and by sklearn library

```python
sk_w = lr_model.coef_
sk_b = lr_model.intercept_

print(f'w => manually: {w_fin:8.4f}  |  sklearn  {sk_w[0]:8.4f}   |  diff: {w_fin - sk_w[0]:8.4f}')
print(f'b => manually: {b_fin:8.4f}  |  sklearn  {sk_b:8.4f}   |  diff: {b_fin - sk_b:8.4f}')
```
